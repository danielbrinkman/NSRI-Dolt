```python
import sqlalchemy
import pandas as pd

engine = sqlalchemy.create_engine("mysql+mysqlconnector://root:secret2@0.0.0.0:3307/edb")
```

```python
pd.read_sql("SHOW TABLES;", engine)
```

```python
users = [
    {
        'username':'dusty',
        'password': 'secret1',
        'p_hash': hash('secret1')
    },
    {
        'username':'dan',
        'password': 'secret2',
        'p_hash': hash('secret2')
    },
    {
        'username':'kirk',
        'password': 'secret3',
        'p_hash': hash('secret3')
    }
]

users = pd.DataFrame(users)
```

```python
users.to_sql("users", con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql("SHOW TABLES;", engine)
```

```python
user = 'dusty'
password = 'secret1'

pword_match = pd.read_sql(f"SELECT p_hash FROM users WHERE username='{user}'", engine).values.ravel()[0]==hash(password)

print(pword_match)
```
